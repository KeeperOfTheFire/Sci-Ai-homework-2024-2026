```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0, xf = 0, 2
dx = 0.05
num_steps = int(xf/dx) + 1

x = np.linspace(x0, xf, num_steps)
#x = np.arange(x0, xf+.01, dx)



A = np.zeros((num_steps, num_steps))
b = np.zeros_like(x)

for i, xi in enumerate(x):
    if i == 0:
        A[0, 0] = 1
        b[0] = 1
    elif i == num_steps-1:
        A[-1, -1] = 1
        b[-1] = 5
    else:
        A[i, i - 1] = 1
        A[i, i] = -2 - dx**2
        A[i, i + 1] = 1
        b[i] = 2*xi*dx**2 
        
y = np.linalg.solve(A, b)
plt.plot(x, y, 'o-')
plt.show()
```

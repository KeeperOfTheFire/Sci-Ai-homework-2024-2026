```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def fun(t, y):
    dy = y**2 - 4*y + 3
    return dy

t0 = 0
tf = 2
t_step = 500
t_span = (t0, tf)
y0 = [0]

linspace = np.linspace(t0, tf, t_step)

sol = solve_ivp(fun, t_span, y0, t_eval=linspace)


plt.figure(1)
plt.plot(sol.t, sol.y[0])
plt.show()
```

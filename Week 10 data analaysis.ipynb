```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path= "/Users/mratcliff/Desktop/SciAi/Week 10/StudentPerformanceFactors.csv"

df = pd.read_csv(path)

#print(df.columns)

#convert String data to integers
df['Parental_Involvement'] = df['Parental_Involvement'].map({'Low': 0, 'Medium': 1, 'High': 2})
df['Access_to_Resources'] = df['Access_to_Resources'].map({'Low': 0, 'Medium': 1, 'High': 2})
df['Extracurricular_Activities'] = df['Extracurricular_Activities'].map({'No': 0, 'Yes': 1,})
df['Motivation_Level'] = df['Motivation_Level'].map({'Low': 0, 'Medium': 1, 'High': 2})
df['Internet_Access'] = df['Extracurricular_Activities'].map({'No': 0, 'Yes': 1,})
df['Family_Income'] = df['Family_Income'].map({'Low': 0, 'Medium': 1, 'High': 2})
df['Teacher_Quality'] = df['Teacher_Quality'].map({'Low': 0, 'Medium': 1, 'High': 2})
df['School_Type'] = df['School_Type'].map({'Public': 0, 'Private': 1,})
df['Peer_Influence'] = df['Peer_Influence'].map({'Negative': 0, 'Neutral': 1, 'Positive': 2})
df['Learning_Disabilities'] = df['Learning_Disabilities'].map({'No': 0, 'Yes': 1,})
df['Parental_Education_Level'] = df['Parental_Education_Level'].map({'High School': 0, 'College': 1, 'Postgraduate': 2})
df['Distance_from_Home'] = df['Distance_from_Home'].map({'Near': 0, 'Moderate': 1, 'Far': 2})
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1,})


#df2 = df[['Attendance', "Hours_Studied", "Sleep_Hours", "Previous_Scores", "Tutoring_Sessions", "Physical_Activity", "Exam_Score", "Parental_Involvement"]]

correlation_matrix = df.corr()
threshold = 0.1 #sets the correlation threshold

filtered_corr = correlation_matrix.where(correlation_matrix.abs() > threshold) #sets filtered correlation matrix to be the values of the correlation matrix where above the threshold

filtered_corr = filtered_corr.dropna(thresh=2, axis=0).dropna(thresh=2, axis=1) #removes columns with only 1 collelated value (the diagonal) resulting in only the columns with a corelation left over

highest_correlation = filtered_corr['Exam_Score'].sort_values(ascending=False)

print(highest_correlation[1:])

print(f'The highest correlation to exam score is: {highest_correlation.index[1]} with a correlation of {highest_correlation.values[1]}')


sns.heatmap(filtered_corr, vmin=0, vmax=1, annot=True)
plt.show()
sns.lineplot(x=df['Exam_Score'], y=df['Attendance'])

```

From the Heatmap we see that the highest correlation to exam score is Attendance followed by hours studied, followed by previous scores
